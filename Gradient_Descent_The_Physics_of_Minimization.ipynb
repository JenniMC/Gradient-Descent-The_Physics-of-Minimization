# Gradient Descent

Gradient Descent is a popular optimization algorithm in Data Science. Its main goal is to find the minimun value of a funtion. We use it to train machine learning models by reducing the "cost" (the error)



1.   The Cost Funtion: This function measures how "wrong" the model is. Our goal is to make this value as close to zeros as possible.
2.   The Gradient: This is a vector of partial derivatives. It tells us the slope of the function at a specific point.
3. Learning Rate($\alpha$): This parameter controls how big our steps are. If the rate is too high, we migt overshoot the minimum. If it is too low, the algorithm takes too long to finish.



The Gradient Descent is
$$w_{i+1}=w_i-\eta \frac{df}{dw}(w_i)$$
where $\eta$ is learning rate

## The Physics Problem:Finding the Lowest Point
Imagine a metal ball rolling on a complex, uneven landscape (like a valley). Gravity pulls the ball downwards. In this scenario:
1. The Landscape is our Cost Function:
$f(x,y)=x^2+y^2$
2. The Gravity is our Gradient: It always pulls in the direction of the steepest descent
3. The Minimum is the "hole" at the bottom where the ball stops (the most stble state)



The objective funtion is

```python
def f(x): return x**2 + y**2
```

The derivative of function is

```python
def gradient(x,y): return 2*x,2*y
```

We need the parameters, x, learning rate and iterations

```python
x,y=5,5
learning_rate=0.1
iteractions=20
path_x, path_y=[x],[y]
```

So, the Gradient Descent loop

```python
for i in range(iteractions):
  dx,dy=gradient(x,y)
  x=x-learning_rate*dx
  y=y-learning_rate*dy
  path_x.append(x)
  path_y.append(y)
```

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection='3d')
X,Y=np.meshgrid(np.linspace(-6,6,20),np.linspace(-6,6,20))
Z=X**2+Y**2

def f(x, y): return x**2 + y**2

ax.plot_surface(X,Y,Z,alpha=0.3, cmap='viridis')
ax.scatter(path_x, path_y, [f(a,b) for a,b in zip(path_x, path_y)], color='red', marker='o')
plt.title('Gradient Descent: The Physics of Minimization')
plt.show()
```
